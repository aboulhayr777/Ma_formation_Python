## 🧠 Théorie

### 1 La fonction `print()`
- Sert à **afficher des informations à l’écran**
- Peut afficher du texte, des nombres, des variables
- Permet de **rendre le programme compréhensible** pour l’utilisateur 

```python
# Exemples :   
print("Bonjour !")  
nom = "Ali"  
print("Nom :", nom) 
```

```python
age = 20
print(f"Nom : {nom}, Age : {age}")
```

### 2 La fonction input()

- Sert à **demander une information à l’utilisateur**
- Retourne **toujours une chaîne de caractères (str)**
- Si on veut faire des calculs, **il faut convertir le type (int(), float())**

```python
ville = input("Votre ville :")
print(type(ville))
print(f"Ville : {ville}.")
```

```python
nom = input("Entrez votre nom : ")
print(type(nom))
print(f"Bonjour {nom} !")
```

```python
age = input("Entrez votre âge : ")
print(type(age))
print(f"Vous avez {age} ans")
```

### Il est important de convertir les variables saisies en int ou float avant de réaliser des calculs arithmètiques sinon celles-ci sont considerées comme des chaînes de caractères (str).
#### Observons les deux codes suivants qui est presque le même code mais avec des rresultats différents :

### Code 1 : Sans conversion

```python
x = input("Donne un premier nombre : ")
y = input("Donne un deuxième nombre : ")

print(type(x))
print(type(y))
print(f"La somme est {x+y}.")
```

### Code 2 : Avec conversion

```python
x = input("Donne un premier nombre : ")
y = input("Donne un deuxième nombre : ")

x_int = int(x)
y_int = int(y)

print("x", type(x))
print("y", type(y))

print("x_int", type(x_int))
print("y_int", type(y_int))

print(f"(1) La somme est {x+y}.")
print(f"(2) La somme est {x_int+y_int}.")
```
