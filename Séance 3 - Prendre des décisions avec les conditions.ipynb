## 1- Comparaisons
Avec **python**, on utilise les symboles suivants pour faire des comparaisons :  
- `==` : **égal**
- `!=` : **différent**
- `>` et `>=` : **supérieur** et **supérieur ou égal**
- `<` et `<=` : **inférieur** et **inférieur ou égal**

**_Remarque_** : _A chaque comparaison, il y aura une réponse de type bool **(True/False)**._
- **True** si la condition est vérifiée.
- **False** si la condition est non vérifiée.

```python
age = 18
```

```python
print(age == 18)
```

```python
print(age == 17)
```

```python
print(age != 18)
```

```python
print(age != 10)
```

```python
age > 18
```

```python
age < 30
```

## 2- Structure conditionnelle `if`, `elif` et `else`:
**if** signifie **si**, **elif** signifie **sinon si** et **else** signifie **sinon**.

```python
age = 15

if age >= 18 :
    print("Vous êtes majeur.")
else : 
    print("Vous êtes mineur.")
```

```python
note = 14

if note >= 16:
    print("Très bien")
elif note >= 12:
    print("Bien")
else:
    print("Insuffisant")
```

## 3- Utilisation de `and` et `or` avec les conditions

```python
jour = ""

match jour:
    case "lundi":
        print("Début de semaine")
    case "vendredi":
        print("Presque weekend")
    case _:
        print("Jour normal")
```
