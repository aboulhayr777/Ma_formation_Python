# A- Les variables:

## 1- Qu'est-ce qu'une variable en python ?

En python, une **variable** est un **nom** qu'on donne à une **valeur** pour la retenir en mémoire.

### Exemple:
**En math, on dit par exemple X=5. Cela veut dire qu'on donne à la lettre X la valeur de 5.**

**En python, on écrit:
X=5**

```python
# Déclaration de variables simples
revenu = 50000
age = 30
note = 15.5
nom = "Fatou"
```

### print(): à quoi ça sert ?
En python, la fonction **print()** sert à afficher un resultat l'écran.

Ces lignes veulent dire:

- la variable *revenu* contient 50000.
- la variable *age* contient 30.
- la variable *note* contient 15.5.
- la variable *nom* contient "Fatou".

### Affichons nos variables

```python
print(revenu)
```

```python
print(age)
```

```python
print(note)
```

```python
print(nom)
```

## 2- Les types de variables en python
Chaque variable a un type de données. 

Voici les principaux types utilisés en analyse:

- **Nombre entier** appelé **int** en python: `age = 25`
- **Nombre décimal** appelé **float** en python: `note = 17.5`
- **Texte** appelé **str** en python: `ville = "Conakry"`
- **Vrai ou Faux** appelé **bool** en python: `verite = True` ou `verite = False`

**Pour vérifier le type d'une variable, on utilise la fonction type(*variable*).**

```python
# Utilisation de la fonction type().
revenu = 50000
print(type(revenu))
```

```python
nom = "Moussa"
print(type(nom))
```

```python
note = 12.5
print(type(note))
```

```python
verite = True
print(type(verite))
```

# B- Opérations sur les variables:

```python
# Création des variables
a = 5
b = -7
c = 2.5
d = 1/2
```

```python
print(a+b)
```
