## 2- Les types de variables en python
Chaque variable a un type de données. 

Voici les principaux types utilisés en analyse:

- **Nombre entier** appelé **int** en python: `age = 25`
- **Nombre décimal** appelé **float** en python: `note = 17.5`
- **Texte** appelé **str** en python: `ville = "Conakry"`
- **Vrai ou Faux** appelé **bool** en python: `verite = True` ou `verite = False`

**Pour vérifier le type d'une variable, on utilise la fonction type(*variable*).**

```python
# Utilisation de la fonction type().
revenu = 50000
print(type(revenu))
```

```python
nom = "Moussa"
print(type(nom))
```

```python
note = 12.5
print(type(note))
```
